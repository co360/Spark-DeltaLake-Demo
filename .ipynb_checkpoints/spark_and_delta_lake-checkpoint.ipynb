## Ingest Raw Data Into Data Lakehouse

### Gather Raw Data From Data Store

```python
# define function to handle downloading a file from a given url
import requests 
def download_url(url, save_path, chunk_size = 128):
    r = requests.get(url, stream = True)
    with open(save_path, 'wb') as fd:
        for chunk in r.iter_content(chunk_size = chunk_size):
            fd.write(chunk)
```

```python
# download zip file from website 
# normally, data would probably be stored in elastic cloud storageâ€•e.g., AWS S3
download_url("https://github.com/JoeKnittel/joeknittel.github.io/raw/main/tidymodels_data.zip", "data.zip")
```

```python
# unzip zip file
import zipfile
with zipfile.ZipFile("./data.zip", 'r') as zip_ref:
    zip_ref.extractall("./")
```

### Create a SparkSession with Delta Lake Enabled

```python
# this sets up a spark connection in local mode, which we'll use for our simple single-node example
# for different cluster configurations (e.g., YARN mode), adjust this cell accordingly
# credentials would also be required here for ingesting files from cloud storage  
import pyspark
from delta import *
builder = pyspark.sql.SparkSession.builder.appName("DeltaTest") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")
spark = configure_spark_with_delta_pip(builder).getOrCreate()
```

### Ingest Raw Data into Spark DataFrame

```python
# ingests CSV file we just downloaded into spark dataframe
# if we were using AWS S3--given appropriate credentials--just change to '.load("<S3 Bucket URL>")'
df = (spark.read.format("csv")
        .option("inferSchema", "true")
        .option("header", "true")
        .load("./Data/data.csv"))
```

### Write Spark DataFrame Contents Into a Bronze Delta Table

```python
df.write.format("delta").mode("overwrite").save("./delta/bronze/raw")
```

### Load Delta Table Back Into Spark and Inspect Its Contents

```python
data = spark.read.format("delta").load("./delta/bronze/raw")
data.select(data.columns[1:10]).show(10)
```

## Model Pipeline

### Split Data

```python
trainDF, testDF = data.randomSplit([.80, .20], seed = 100)
```

```python
print("Train:", trainDF.count(), "\nTest: ", testDF.count())
```

### Convert Spark DataFrame Into Pandas-on-Spark DataFrame

```python
import os
os.environ["PYARROW_IGNORE_TIMEZONE"] = "1"
```

```python
# we can utilize all the familiar pandas functionality 
# while operating on spark dataframes 
pd_trainDF = trainDF.to_pandas_on_spark()
pd_trainDF.head(10)
```

### Identify Columns to be Omitted

```python
# identify columns with >40% missing values
many_missing = pd_trainDF.isna().sum()/len(pd_trainDF) > 0.4
missing = list(pd_trainDF.loc[:,many_missing.values].columns)
missing
```

```python
# identify categorical vars with many levels
pd_trainDF[categorical].nunique()  
```

```python
# identify predictors
disregard = missing + ['Id', 'Medical_History_2', 'Response']
predictors = [x for x in pd_trainDF.columns if x not in disregard]
```

### Generate a Hashed Feature Vector of Remaining Columns

```python
# this does one-hot encoding on categorical variables and generates
# a sparse feature vector, which improves computation efficiency 
from pyspark.ml.feature import FeatureHasher
hasher = FeatureHasher(inputCols = predictors,
                       outputCol = "hashed_features")
```

### Discard Zero-Variance Features

```python
from pyspark.ml.feature import VarianceThresholdSelector
selector = VarianceThresholdSelector(featuresCol = hasher.getOutputCol(), outputCol = "features")
```

### Random Forest Model Spec

```python
from pyspark.ml.classification import RandomForestClassifier
rf = RandomForestClassifier(labelCol = 'Response', seed = 200)
```

### Hyperparameter Tuning Grid

```python
from pyspark.ml.tuning import ParamGridBuilder
paramGrid = (ParamGridBuilder()
            .addGrid(rf.maxDepth, [2, 4, 6])
            .addGrid(rf.numTrees, [50, 150])
            .build())
```

### Evaluation Metric

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
evaluator = MulticlassClassificationEvaluator(labelCol = "Response",
                                              predictionCol = "prediction",
                                              metricName = "accuracy")
```

### Cross-Validation Process

```python
from pyspark.ml.tuning import CrossValidator
cv = CrossValidator(estimator = rf,
                    evaluator = evaluator,
                    estimatorParamMaps = paramGrid,
                    numFolds = 10,
                    parallelism = 4,
                    seed = 300)
```

### Construct the Model Pipeline

```python
from pyspark.ml import Pipeline
pipeline = Pipeline(stages = [hasher, selector, cv])
```

### Fit the Model on Training Data Using the Pipeline

```python
# performs 10-fold cross-validation to identify best set of hyperparameters, 
# then fits the model on the entire training set
pipelineModel = pipeline.fit(trainDF)
```

## Model Evaluation

### Predictions on Testing Set

```python
predDF = pipelineModel.transform(testDF)
predDF.select("features", "Response", "prediction").show(10)
```

### Evaluate Accuracy on Testing Set

```python
evaluator.evaluate(predDF)
```

### Confusion Matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
predictions = predDF.select("Response", "prediction").to_pandas_on_spark()
r = predictions['Response'].to_numpy()
p = predictions['prediction'].to_numpy()
cm = confusion_matrix(r,p)
ax = sns.heatmap(cm, annot = True, fmt = 'g', 
                 cmap = sns.color_palette("ch:s = -.2, r = .6", as_cmap = True))
ax.set_xlabel('Predicted Response')
ax.set_ylabel('Actual Response');
ax.xaxis.set_ticklabels(range(1,9))
ax.yaxis.set_ticklabels(range(1,9))
plt.show()
```
